# Data Preprocessing

```python
import pandas as pd
import datetime as dt
import numpy as np

# Recency
df_recency = df.groupby(by='ID_Donatur', as_index=False)['Tanggal'].max()
df_recency.columns = ['ID_Donatur', 'LastPurchaseDate']
recent_date = df_recency['LastPurchaseDate'].max()
df_recency['Recency'] = df_recency['LastPurchaseDate'].apply( lambda x: (recent_date - x).days)
display(df_recency)
print("--------------------------------------------------")

# Frequency
frequency_df = df.drop_duplicates().groupby( by=['ID_Donatur'], as_index=False)['Tanggal'].count()
frequency_df.columns = ['ID_Donatur', 'Frequency']
display(frequency_df)
print("--------------------------------------------------")

# Monetary 
monetary_df = df.groupby(by='ID_Donatur', as_index=False)['Nominal'].sum()
monetary_df.columns = ['ID_Donatur', 'Monetary']
display(monetary_df)
print("---------------------------------------------------")

# RFM
rf_df = df_recency.merge(frequency_df, on='ID_Donatur')
rfm_df = rf_df.merge(monetary_df, on='ID_Donatur').drop(columns='LastPurchaseDate')
display(rfm_df)
```

```python
rfm_df.info()
```

# Analisis



```python
rfm_df['R_rank'] = rfm_df['Recency'].rank(ascending=False)
rfm_df['F_rank'] = rfm_df['Frequency'].rank(ascending=True)
rfm_df['M_rank'] = rfm_df['Monetary'].rank(ascending=True)
 
# normalizing the rank of the customers
rfm_df['R_rank_norm'] = (rfm_df['R_rank']/rfm_df['R_rank'].max())*100
rfm_df['F_rank_norm'] = (rfm_df['F_rank']/rfm_df['F_rank'].max())*100
rfm_df['M_rank_norm'] = (rfm_df['F_rank']/rfm_df['M_rank'].max())*100
 
rfm_df.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)
 
rfm_df.head()
```

```python
# Rumus yang digunakan untuk menghitung skor rfm adalah : 
# 0.15*Recency score + 0.28*Frequency score + 0.57*Monetary score
```

```python
rfm_df['RFM_Score'] = 0.15*rfm_df['R_rank_norm']+0.28 * \
    rfm_df['F_rank_norm']+0.57*rfm_df['M_rank_norm']
rfm_df['RFM_Score'] *= 0.05
rfm_df = rfm_df.round(2)
rfm_df[['ID_Donatur', 'RFM_Score']]
```

```python
rfm_df.info()
```

# Segmentasi Donatur

```python
# KLASIFIKASI SEGMEN
# rfm score >4.5 : Top Customer
# 4.5 > rfm score > 4 : High Value Customer
# 4>rfm score >3 : Medium value customer
# 3>rfm score>1.6 : Low-value customer
# rfm score<1.6 :Lost Customer
```

```python
rfm_df["Customer_segment"] = np.where(rfm_df['RFM_Score'] > 4.5, "Top Customers",
                            (np.where(rfm_df['RFM_Score'] > 4,"High value Customer",
                            (np.where(rfm_df['RFM_Score'] > 3, "Medium Value Customer",
                            np.where(rfm_df['RFM_Score'] > 1.6,'Low Value Customers',
                            'Lost Customers'))))))
rfm_df[['ID_Donatur', 'RFM_Score', 'Customer_segment']]
```
